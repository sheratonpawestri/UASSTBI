```python
import pandas as pd
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
import time
```

```python
dataset = pd.read_excel('dataset.xlsx')
dataset
```

```python
def check(query):
    for i in range(len(dataset)):
        penyusun_aksara = dataset['aksara_dasar'][i].split(",")
        if query in penyusun_aksara: 
            aksara = dataset['data'][i]
            output(aksara)
```

```python
def output(aksara):
    path = "C:/Users/ASUS/Documents/Tugas STBI/dataset/"+aksara+".jpg"
    img = cv2.imread(path)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    print("Aksara di atas dibaca: ", aksara)
    print("_________________________________________________________")
```

```python
def run():
    masukkan = input ("Masukkan satu huruf aksara Jawa: ")
    masuk = masukkan.lower()
    start_time = time.time()
    check(masuk)
    return ("Waktu pencarian: %s detik" % (time.time()-start_time))
```

```python
run()
```
